## 2.   Preprocessing Data or Clean Data

> Converting data from label to range number

```python
import pandas as pd

## import Preprocessing to call function LabelEncoder to covert data
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder for each categorical column
label_encoder = LabelEncoder()
# Fit and transform the 'Gender' column to numerical values
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['education'] = label_encoder.fit_transform(df['education'])
df['prevalentStroke'] = label_encoder.fit_transform(df['prevalentStroke'])
df['Heart_ stroke'] = label_encoder.fit_transform(df['Heart_ stroke'])
```

```python
def feature_selection(num):
  correlation_matrix = X.corr()
  # You can use a threshold to decide which features to keep or remove
  # print(correlation_matrix.iloc[9,1])
  highly_correlated = set()
  for j in range(len(correlation_matrix)):
      if abs(correlation_matrix.iloc[num,j]) > 0.05:
          colname = correlation_matrix.columns[j]
          highly_correlated.add(colname)
  return highly_correlated;

print(feature_selection(9))

correlation_matrix = df.corr()['Heart_ stroke']
top_10_features = correlation_matrix.abs().nlargest(10).index
top_10_corr_values = correlation_matrix[top_10_features]

plt.figure(figsize=(10, 11))
plt.bar(top_10_features, top_10_corr_values)
plt.xlabel('Features')
plt.ylabel('Correlation with Target')
plt.title('Top 10 Features with Highest Correlation to Target')
plt.xticks(rotation=45)
plt.show()
# X = X.drop(columns=highly_correlated)
```

> Split data for Training and Testing 70% and 30%  

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```


## 5.   Training Model with another Algorithim



*   Logistic Regression
*   Decision Tree

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# Initialize the Logistic Regression model
model = LogisticRegression()
# Train the model
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)
print(y_pred)
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)


table_accuracy ={"Logistic Regression":accuracy}
table_accuracy = pd.Series(table_accuracy, name='cigsPerDay')
table_accuracy = pd.DataFrame(table_accuracy)

# Display results
print("Accuracy: {accuracy}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```
